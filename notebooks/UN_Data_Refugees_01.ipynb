```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Read and clean data
refugees_df = pd.read_csv("../data/UNdata_refugees.csv", skipfooter=3, engine='python')
ref_df = refugees_df[["Country or territory of asylum or residence", "Country or territory of origin", 
                      "Year", "Refugees<sup>*</sup>"]]
ref_df = ref_df.rename({"Country or territory of asylum or residence": "asylum", 
                        "Country or territory of origin": "origin", "Refugees<sup>*</sup>": "refugees"}, axis="columns")
ref_df["Year"] = ref_df["Year"].astype(int)
```

```python
plt.plot(ref_df.groupby("Year").sum())
plt.title("Total Refugees Outflows by Year 1975-2016 (in millions)")
print() 
```

```python
# Number of refugees outflow sorted by country
ref_df[["origin", "refugees"]][ref_df["origin"] != "Various"].groupby("origin").sum()\
    .sort_values("refugees", ascending=False).head(20)
```

```python
# Number of refugees taken in sorted by country
ref_df[["asylum", "refugees"]].groupby("asylum").sum().\
    sort_values("refugees", ascending=False).head(20)
```

```python
ref_df[ref_df["origin"] == "Various"].sort_values("refugees", ascending=False).head(20)
```

```python
ref_df.groupby(["origin","asylum"]).sum()
```

```python
ref_df[(ref_df["origin"] == "Afghanistan") & (ref_df["refugees"] > 1000) & 
       (ref_df["Year"] == 2016)].sort_values("refugees", ascending=False)
```

```python
plt.plot(ref_df[(ref_df["origin"] == "Afghanistan") & (ref_df["refugees"] > 1000)].\
         sort_values("refugees", ascending=False)[["Year","refugees"]].\
        groupby("Year").sum().sort_values("Year"))
```

```python
plt.plot(ref_df[(ref_df["origin"] == "Iraq") & (ref_df["refugees"] > 1000)].\
         sort_values("refugees", ascending=False)[["Year","refugees"]].\
        groupby("Year").sum().sort_values("Year"))
```
